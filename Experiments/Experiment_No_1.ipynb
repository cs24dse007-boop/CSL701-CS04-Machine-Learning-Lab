```python
# Task 1: Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 2: Load train.csv into a Pandas DataFrame
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
# Task 3: Display the first 10 rows
print(df.head(10))
```

```python
# Task 4: Display the last 5 rows
print(df.tail(5))
```

```python
# Task 5: Check dataset dimensions
print("Dataset shape:", df.shape)
# Expected output: (891, 12)
```

```python
# Task 6: Display dataset information
print(df.info())
```

```python
# Task 7: Generate summary statistics
print(df.describe())

# Record mean, minimum, and maximum values of Age and Fare
print("\nAge Statistics:")
print("Mean:", df["Age"].mean())
print("Minimum:", df["Age"].min())
print("Maximum:", df["Age"].max())

print("\nFare Statistics:")
print("Mean:", df["Fare"].mean())
print("Minimum:", df["Fare"].min())
print("Maximum:", df["Fare"].max())
```

```python
# Task 8: Identify missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Task 9: Fill missing Age values with the median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Verify that missing Age values are filled
print("Missing Age values:", df["Age"].isnull().sum())
```

```python
# Task 10: Filter surviving passengers (Survived == 1)
survivors = df[df["Survived"] == 1]

# Display the first few surviving passengers
print(survivors.head())
```

```python
# Task 11: Convert Fare column to a NumPy array
fare_array = df["Fare"].to_numpy()

# Display the NumPy array
print(fare_array)
```

```python
# Task 12: Calculate maximum and average Fare
max_fare = np.max(fare_array)
avg_fare = np.mean(fare_array)

print("Maximum Fare:", max_fare)
print("Average Fare:", avg_fare)
```

```python
# Task 13: Plot histogram of Age
plt.figure(figsize=(8, 5))
plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Task 14: Plot bar chart of Survival Count
plt.figure(figsize=(6, 4))
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Task 15: Plot scatter graph of Age vs Fare
plt.figure(figsize=(8, 5))
plt.scatter(df["Age"], df["Fare"], alpha=0.6, color="green")
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
